```python
def getByCategory(categoryID,length,filename):
    response = requests.get(base_url+page_url+str(1)+category_url+str(categoryID))
    resJson = response.json()
    last_page = resJson['data']['last_page'];
    if(length > last_page):
        print("Длина большая беру "+str(last_page))
        length = last_page
    data = pd.DataFrame(resJson['data']['data'])[columns]
    for x in range(2,length):
        time.sleep(10)
        print(x)
        response = requests.get(base_url+page_url+str(x)+category_url+str(categoryID))
        resJson = response.json()
        df = pd.DataFrame(resJson['data']['data'])[columns]
        data=data.append(df)
        df.to_json('./pages/'+filename+'-'+str(x)+'.json')
    data=data.reset_index(drop=True)
    data.to_json(filename+'.json')
    return data
```
